```python
data = data.drop("CommentId", axis=1)
data = data.drop("VideoId", axis=1)
data= data.drop("IsHatespeech", axis=1)
data = data.drop("IsAbusive", axis=1)
data = data.drop("IsThreat", axis=1)
data = data.drop("IsProvocative", axis=1)
data = data.drop("IsObscene", axis=1)
data = data.drop("IsRacist", axis=1)
data = data.drop("IsNationalist", axis=1)
data = data.drop("IsSexist", axis=1)
data = data.drop("IsHomophobic", axis=1)
data = data.drop("IsReligiousHate", axis=1)
data = data.drop("IsRadicalism", axis=1)
```

```python
data.head(5)
```

```python
data.rename(columns={'Text': 'comment'}, inplace=True)
data.rename(columns={'IsToxic': 'toxic'}, inplace=True)
```

```python
data['toxic'] = data['toxic'].replace(False, 0)
data['toxic'] = data['toxic'].replace(True, 1)
```

```python
import nltk
nltk.download('punkt')
nltk.download('omw-1.4')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
from nltk import WordNetLemmatizer
from nltk import pos_tag, word_tokenize
from nltk.corpus import stopwords as nltk_stop
from nltk.corpus import wordnet
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
import re
```

```python
def prepare_text(text): #prepare_text function credit goes to AiAnytime, code used to increase productivity time.
    def get_wordnet_pos(treebank_tag):
        if treebank_tag.startswith('J'):
            return wordnet.ADJ
        elif treebank_tag.startswith('V'):
            return wordnet.VERB
        elif treebank_tag.startswith('N'):
            return wordnet.NOUN
        elif treebank_tag.startswith('R'):
            return wordnet.ADV
        else:
            return wordnet.NOUN
    text = re.sub(r'[^a-zA-Z\']', ' ', text)
    text = text.split()
    text = ' '.join(text)
    text = word_tokenize(text)
    text = pos_tag(text)
    lemma = []
    for i in text: lemma.append(wordnet_lemmatizer.lemmatize(i[0], pos = get_wordnet_pos(i[1])))
    lemma = ' '.join(lemma)
    return lemma
```

```python
data['lemmatized_comment'] = data['comment'].apply(lambda x: prepare_text(x))
```

```python
data.head(30)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split 
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
```

```python
corpus = data['lemmatized_comment'].values.astype('U')
```

```python
stopwords = list(set(nltk_stop.words('english')))
```

```python
tfidf_count = TfidfVectorizer(stop_words = stopwords)
tfidf = tfidf_count.fit_transform(corpus)
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    tfidf, data['toxic'], test_size = 0.7, random_state= 23, shuffle=True
)
```

```python
bayes_model = MultinomialNB()
```

```python
bayes_model = bayes_model.fit(x_train, y_train)
```

```python
predicted_proba_y = bayes_model.predict_proba(x_test)[::, 1]
```

```python
predicted_proba_y 
```

```python
final_roc_auc = roc_auc_score(y_test, predicted_proba_y )
```
